```python
# Orders Table - Structure and Sample Data
print("📋 ORDERS TABLE STRUCTURE")
print("="*60)

# Show schema
orders_df = spark.table("workspace.lesson_db.orders")
orders_df.printSchema()

print("\n📊 SAMPLE DATA (first 5 rows)")
print("="*60)

# Display first few rows in a nice format
display(orders_df.limit(5))
```

```python
# Customers Table - Structure and Sample Data
print("👥 CUSTOMERS TABLE STRUCTURE")
print("="*60)

# Show schema
customers_df = spark.table("workspace.lesson_db.customers")
customers_df.printSchema()

print("\n📊 SAMPLE DATA (first 5 rows)")
print("="*60)

# Display first few rows in a nice format
display(customers_df.limit(5))
```

```python
display(customers_df.agg({"customer_id": "count"}).withColumnRenamed("count(customer_id)", "customer_count"))
```
